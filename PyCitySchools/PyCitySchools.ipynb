```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools.csv"
student_data_to_load = "students.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Dropping any N/A values within the data set
school_data_complete = school_data_complete.dropna(how='any')

#Displaying our data set 
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_df = school_data_complete.groupby("school_name")

school_count_df = school_df.nunique()["school_name"].sum()

school_money_df = school_data['budget'].sum()

student_count_df = school_df['Student ID'].value_counts().sum()

math_score_df = round(school_data_complete['math_score'].mean(),6)

rding_score_df = round(school_data_complete['reading_score'].mean(),6)

pass_math_df = school_data_complete.loc[school_data_complete['math_score'] >= 70]

pass_read_df = school_data_complete.loc[school_data_complete['reading_score'] >= 70]


ind_math_score_df = round(len(pass_math_df) / student_count_df * 100,6)

ind_read_score_df = round(len(pass_read_df) / student_count_df * 100,6) 


perc_overall = len(school_data_complete.loc[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >= 70)])/student_count_df*100


District_Analysis_df = {"Total Schools": [school_count_df],
                            "Total Students": [student_count_df],
                            "Total Budget": [school_money_df],
                            'Average Math Score':[math_score_df],
                            'Average Reading Score':[rding_score_df],
                            "% Passing Math":[ind_math_score_df],
                            "% Passing Reading":[ind_read_score_df],
                            "% Pass Overall":[perc_overall]}



District_Analysis_df = pd.DataFrame(District_Analysis_df)

District_Analysis_df = District_Analysis_df.style.format({'Total Students':'{:,}',
                                          'Total Budget':'${0:,.0f}'})
District_Analysis_df 


```
