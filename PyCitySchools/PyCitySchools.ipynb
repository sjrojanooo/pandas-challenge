```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools.csv"
student_data_to_load = "students.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Dropping any N/A values within the data set
school_data_complete = school_data_complete.dropna(how='any')

#Displaying our data set 
school_data_complete
```

```python
#Printing a statistical overview of the data set 
school_data_complete.describe()
```

```python
#Grouping data by the school name with the groupby function
school_df = school_data_complete.groupby("school_name")

#Counting the unique values of school within the data set 
#nunique function ignores duplicate values within the data set
school_count_df = school_df.nunique()['school_name'].sum()

#printing count
school_count_df
```

```python
#Calculating the sum of the total budget for the distric 
district_budget  = school_data['budget'].sum()

#displaying the amount 
district_budget
```

```python
#Grabbing the total count of students students in the district 
student_count_df = school_df['Student ID'].count().sum()

#printing the count
student_count_df
```

```python
#calclating the average math & reading score in the district 
math_score_df = round(school_data_complete['math_score'].mean(),6)

rding_score_df = round(school_data_complete['reading_score'].mean(),6)

print(math_score_df,rding_score_df)

```

```python
#Using the ".loc" statement to locate scores that are 70 and above, and storing them in the corresponding variables 
pass_math_df = school_data_complete.loc[school_data_complete['math_score'] >= 70]

pass_read_df = school_data_complete.loc[school_data_complete['reading_score'] >= 70]

#Dividing the count variables by the total number of students in the district 
#Each calculations is done individually
ind_math_score_df = round(len(pass_math_df) / student_count_df * 100,6)

ind_read_score_df = round(len(pass_read_df) / student_count_df * 100,6) 

#printing the values 
print(ind_math_score_df)

print(ind_read_score_df)
```

```python
#Had some trouble finding the obtaining the correct value for this section. 
#To calculate the overall passing percentage we can't simply use the variables "pass_math_df & pass_read_df"
#Instead we need to perform a .loc statement again but this time with two conditions
#This gives us the precise overall passing percentage in the District 
perc_overall = len(school_data_complete.loc[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >= 70)])/student_count_df * 100


print(perc_overall)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Grabbing the total of students in each school 
stu_per_school = school_data_complete.groupby(['school_name'])['student_name'].count()

#Budget per school 
school_budget = school_data_complete.groupby(['school_name'])['budget'].mean()

#Budget Per student 
student_budget_average = school_budget / stu_per_school

#Grabbing a count of students that earned a passing percentage in math from each school 
mth_per_school = pass_math_df.groupby(["school_name"]).count()["student_name"]

#Grabbing a count of students that earned a passing percentage in reading from each school 
rding_per_school = pass_read_df.groupby(["school_name"]).count()["student_name"]

#calculating the percentage for math 
schl_math_overall = mth_per_school / stu_per_school * 100

#calculating the percentage for reading 
schl_rding_overall = rding_per_school / stu_per_school * 100 

#Here is where I am stuck 
overall_passing_perc = school_data_complete.loc[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >= 70)]

per_school_over = overall_passing_perc.groupby(["school_name"])['student_name'].count() / stu_per_school * 100 

```

```python
#Using the aggregation function to not only find, but also calculate values within my data frame
#First - find the first value listed under the desired column 
#Mean - calculates the mean for the column 


school_summary = school_data_complete.groupby(['school_name']).agg({"type":"first",
                                                                    "size":"first",
                                                                    "budget":"mean",
                                                                    "math_score":"mean",
                                                                   "reading_score":"mean"})


#Converting school_summary into a data frame 
school_summary = pd.DataFrame(school_summary)

#adding new calculated columns into the data frame 
school_summary['% Passing Math'] = schl_math_overall

school_summary['% Passing Reading'] = schl_rding_overall

school_summary['% Passing Overall'] = per_school_over


school_summary.insert(3, 'Per Student Budget',student_budget_average)


#Renaming the column headers within the data frame 
school_summary = school_summary.rename(columns={'type':'School Type','size':'School Size',
                                         'budget':'Total School Budget','math_score':'Average Math Score',
                              'reading_score':'Average Reading Score'})


#Applying style formatting for total budget and per student budget
#This give those values a currency symbol in front of the data 
school_summary_df = school_summary.style.format({'Total School Budget':'${0:,.2f}',
                                          'Per Student Budget':'${0:,.2f}'})


#Displaying the dataframe 
school_summary_df

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Creating bins for the spending ranges 
spending_bins = [0,583.99,628.99,643.99,674.99]

#Creating labels for the bins 
spending_labels = ["<$584","$585-629","$630-644","$645-675"]

#Segmenting and sorting my data values into bins
#Here I am converting the spending per student of each school into specific ranges 
school_summary["Spending Ranges Per Student"] = pd.cut(school_summary["Per Student Budget"],spending_bins,labels=spending_labels)

#Using the groupby method to gather the mean of each subject score outcome depending on the schools size 
#In order to get not lose any data I converted my tuples to list by adding the double brackets 
scores_by_spending = school_summary.groupby(["Spending Ranges Per Student"])[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]].mean()

#Using the dataframe.round() function I was able to give the correct decimal value to each column 
scores_by_spending.round({"Average Math Score":2, "Average Reading Score":2, "% Passing Math":2, "% Passing Reading":2,"% Passing Overall":2})
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Creating bins for the spending ranges 
size_bins = [0,999.99,1999.99,4999.99]

#Creating labels for the bins 
size_labels = ["Small(<1000)","Medium(1000-2000)","Large(2000-5000)"]

#Segmenting and sorting my data values into bins
#Here I am converting the size of each school into specific ranges 
school_summary["Size"] = pd.cut(school_summary["School Size"],size_bins,labels=size_labels)

#Using the groupby method to gather the mean of each subject score outcome depending on the schools size 
#In order to get not lose any data I converted my tuples to list by adding the double brackets 
size_grade_summary = school_summary.groupby(["Size"])[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]].mean()

#Using the dataframe.round() function I was able to give the correct decimal value to each column 
size_grade_summary.round({"Average Math Score":6, "Average Reading Score":6, "% Passing Math":6, "% Passing Reading":6,"% Passing Overall":6})
```
