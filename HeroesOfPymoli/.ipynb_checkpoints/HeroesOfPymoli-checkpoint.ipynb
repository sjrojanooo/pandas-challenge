```python
%matplotlib notebook
import matplotlib as plt
import pandas as pd 
import numpy as np 

# File to Load (Remember to Change These)
purchase_data = "HeroesOfPymoli.csv"

# Read Purchasing File and store into Pandas data frame
purchase = pd.read_csv(purchase_data)

#Displaying the data frame 
purchase
```

## Player Count

* Display the total number of players


```python
#dropping all rows with missing information and setting items equal to new data set without empty values 
items_df = purchase.dropna(how='any')

#Setting a variable equal to the total count of players in the data set 
total_count_df = items_df['SN'].count()

#Creating a bin so that it for a cleaner presentation to the user 
player_bin = {'Total Player Count':[total_count_df]}

#Displaying the data in the new bing 
player_bin = pd.DataFrame(player_bin)

player_bin
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

* Display the summary data frame

```python
#Setting unique value = to Item ID, since this can be classified as a unique purchase 
unique = items_df["Item ID"]

#Setting total purchase = to the complete count of all items 
total_purch = items_df["Item ID"].count()

#Finding the average price 
avg_price = round(items_df["Price"].mean(),2)

#Find the total revenue 
revenue_df = items_df['Price'].sum()


#summarized data table with unique values 
sum_table_df = {'Number of Unique Items': [unique.nunique()],
               'Average Price': [avg_price],
               'Number of Purchases': [total_purch],
               'Total Revenue':[revenue_df]}

#Creating a new data frame so that I can correctly style and format my data 
data_style_df = pd.DataFrame(sum_table_df)

#Found a the style.format function and the key specifiers to add the currency symbols
#to both my Average Price and Total Revenue 
data_style_df = data_style_df.style.format({'Average Price':'${0:,.2f}',
                                          'Total Revenue':'${0:,.2f}'})

#Displaying the results and formatted data 
data_style_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#finding the percentage of male, female, and non-disclosed players 
#getting a full count of unique counts for each description 
gender_df = items_df.groupby("Gender")

#Finding the unique values using the special number or ID information 
gender_count = gender_df.nunique()["SN"]

#normalize=true function 
#It returns the frequency value by dividing the the unique value by the sum of all values
#This gives us the percentage amount of players 
gender_perc_df = items_df['Gender'].value_counts(normalize=True) 


#Creating the Data Frame to diplay the total count of players and the percentage 
#This data frame did return the table in an unsorted format so we want to format the data in descending order 
#(largest to smalled )
gender_summary_df = pd.DataFrame({"Total Count": gender_count, 
                                 "Percentage of Players": gender_perc_df})

#This nifty code sorts our values, and we give it the condition that ascendind is = false
#so by rule that says descending is true 
sort_gender_sum = gender_summary_df.sort_values(['Total Count','Percentage of Players'], ascending = False)

#applying the same format functions but this time with percentage 
sort_gender_sum = sort_gender_sum.style.format({'Percentage of Players':'{:.2%}'})

#displaying the data frame to the user 
sort_gender_sum
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purch_group_df = items_df.groupby("Gender")

purch_data_df = purch_group_df.nunique()["Purchase ID"]

Avg_purch_df = purch_group_df["Price"].mean()

total_value_df = purch_data_df * Avg_purch_df

gender_purchase_total =  total_value_df / gender_count

print(gender_purchase_total)

Gender_purch_analysis_df = pd.DataFrame({'Purchase Count': purch_data_df,
                            'Average Purchase Price': Avg_purch_df,
                            'Total Purchase Value': total_value_df,
                            'Avg Total Purchase Per Person':gender_purchase_total})

Gender_purch_analysis_df = Gender_purch_analysis_df.sort_values(['Purchase Count'], ascending = False)



#applying the same format functions but this time with percentage 
Gender_purch_analysis_df = Gender_purch_analysis_df.style.format({'Average Purchase Price':'${0:,.2f}',
                                               'Total Purchase Value':'${0:,.2f}',
                                               'Avg Total Purchase Per Person':'${0:,.2f}'})


Gender_purch_analysis_df


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Now I have a better idea of the bins to create for my custom data frame 
age_bins = [0,9.90,14.90,19.90,24.90,29.90,34.90,39.90,100]

#Labels for these bins 
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Placing the data series Age Group into a new column of the data frame 
purchase["Age Group"] = pd.cut(purchase["Age"],age_bins,labels=age_labels)

#Creating a group by based on the Age Group I just created 
age_group = purchase.groupby("Age Group")

age_count = age_group["SN"].nunique()

purch_age_sum = age_count.sum()

perc_purch_df = age_count / purch_age_sum

age_dem = pd.DataFrame({"Total Count": age_count,
                       "Percentage of Players":perc_purch_df})

age_dem.index.name = None


age_dem = age_dem.style.format({'Percentage of Players':'{:.2%}'})

age_dem
```
